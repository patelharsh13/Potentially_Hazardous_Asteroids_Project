```python
# add last_obs and per to create future_obs using delta step 1

from datetime import datetime,timedelta
```

```python
# add last_obs and per to create future_obs using delta step 2

def add_days_to_date(original_date_str, days_to_add):
    original_date = datetime.strptime(original_date_str, '%d-%m-%Y')
    delta = timedelta(days = days_to_add)
    future_obs = original_date + delta
    return future_obs.strftime('%d-%m-%Y')
```

```python
# add last_obs and per to create future_obs using delta step 3

al_pha['future_obs'] = al_pha.apply(lambda row: add_days_to_date(row['last_obs'], row['per']), axis=1)
```

```python
print(al_pha)
```

```python
al_pha.info()
```

```python
#extract only years from future_obs

al_pha['future_obs_y'] = pd.DatetimeIndex(al_pha['future_obs']).year
```
